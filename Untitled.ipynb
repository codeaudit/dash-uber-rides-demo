```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('output.csv')
df.drop("Unnamed: 0", 1)
df["Date/Time"] = pd.to_datetime(df["Date/Time"], format="%Y-%m-%d %H:%M:%S")
df.index = df["Date/Time"]
df.drop("Date/Time", 1, inplace=True)
df.drop("Base", 1, inplace=True)
```

```python
df = dd.from_pandas(df, 2)
df.compute()
```

```python
totalList = []
for month in df.compute().groupby(df.index.month.compute()):
    dailyList = []
    for day in month[1].groupby(month[1].index.day):
        dailyList.append(dd.from_pandas(day[1], chunksize=500000))
    totalList.append(dailyList)
    
totalList = dd.from_array(pd.Series(totalList))
totalList.compute()[0][0].compute()
```
